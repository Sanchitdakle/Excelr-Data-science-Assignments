```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
```

# 4 Feature Selection

```python
from sklearn.ensemble import IsolationForest
```

```python
iso = IsolationForest(contamination=0.05)
numerical_features = df.select_dtypes(include=['number'])
outliers = iso.fit_predict(df[['age', 'hours_per_week']])
df = df[outliers != -1]
```

```python
df
```

```python
iso.fit(numerical_features)
```

```python
outlier_predictions = iso.predict(numerical_features)
# Remove outliers 
df_num = numerical_features[outlier_predictions != -1]
```
