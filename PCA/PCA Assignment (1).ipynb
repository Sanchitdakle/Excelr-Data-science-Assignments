# Task 2: Dimensionality Reduction with PCA:

```python
# Standardized the feature
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
scaler=StandardScaler()
scaler
```

```python
df_scaled=scaler.fit_transform(df)
```

```python
#implement PCA
pca=PCA()
pca.fit(df_scaled)
```

```python
# Scree plot
plt.plot(range(1, len(pca.explained_variance_ratio_)+1), pca.explained_variance_ratio_, marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Variance Explained')
plt.show()
```

```python
# cumulative explained variance
cum_var = pca.explained_variance_ratio_.cumsum()
plt.plot(cum_var, marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.show()
```

```python
pca = PCA()
pca.fit(df_scaled)
df_pca = pca.fit_transform(df_scaled)
df_pca
```

```python
optimal_components = np.argmax(cum_var >= 0.95) + 1

pca = PCA(n_components=optimal_components)
df_pca = pca.fit_transform(df_scaled)
```

```python
df_pca
```

# Task 3: Clustering with Original Data:

```python
# kmeans clustering
from sklearn.cluster import KMeans
k=3
kmeans = KMeans(n_clusters=k)
labels = kmeans.fit_predict(df_scaled)
```

```python
labels
```

```python
# Visualize the clusters
plt.scatter(df_scaled[:, 0], df_scaled[:, 1], c=labels)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('K-means Clustering')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Visualize the clusters in 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df_scaled[:, 0], df_scaled[:, 1], df_pca[:, 2], c=labels)
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
plt.title('K-means Clustering')
plt.show()
```

```python
# Evaluate clustering
from sklearn.metrics import silhouette_score, davies_bouldin_score
print("Silhouette Score:", silhouette_score(df_scaled, labels))
print("Davies-Bouldin Index:", davies_bouldin_score(df_scaled, labels))
```

#Analyze the Impact of PCA:

Potential Benefits:

Noise Reduction:
 High-dimensional data often contains noise or irrelevant features that can obscure the true underlying clusters. Dimensionality reduction can filter out noise, making the clusters more distinct and easier to identify.
Improved Efficiency:
 Clustering algorithms can be computationally expensive for high-dimensional data. Reducing the number of dimensions can lead to faster computation and reduced memory usage.
Mitigation of the Curse of Dimensionality:
 In high-dimensional spaces, the distance between data points becomes less meaningful, hindering distance-based clustering algorithms. Dimensionality reduction can address this issue by projecting data onto a lower-dimensional space where distances are more relevant.

#Analyze trade_offs



Benefits of using PCA before clustering:

Noise reduction: PCA can help remove noise and irrelevant features, potentially leading to more accurate clustering results.
Improved efficiency: Clustering algorithms can run faster on the reduced dataset from PCA.
Curse of dimensionality mitigation: PCA can help address issues caused by high dimensionality, improving the performance of distance-based clustering methods.
Drawbacks of using PCA before clustering:

Information loss: When you reduce dimensions, you might lose some information that could be important for clustering.
Interpretability: The principal components are linear combinations of original features and may be harder to interpret in the context of the original data.
Not always beneficial: In some cases, PCA might not improve clustering results and could even worsen them if important information is lost.
